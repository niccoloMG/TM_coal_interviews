```python
#from https://nicharuc.github.io/topic_modeling/#topic=0&lambda=0&term=
df_topics_lambda = pd.DataFrame(list(range(0, var_no_topics)), columns=['ori_topic_num']) #put all lists into a dataframe

num_terms = 15 # Adjust number of words to represent each topic
for x in [1, 0.8, 0.6, 0.5, 0.4, 0.2, 0]:
    lambd = x
    for i in range(0,var_no_topics): #Adjust this to reflect number of topics chosen for final LDA model
        topics_lambda = list()
        topic = vis.topic_info[vis.topic_info.Category == 'Topic'+str(i+1)].copy()
        topic['relevance'] = topic['loglift']*(1-lambd)+topic['logprob']*lambd
        topics_lambda.append(topic.sort_values(by='relevance', ascending=False).Term[:num_terms].values)
        topics_lambda_list = ', '.join([k for k in list(topics_lambda[0])])
        df_topics_lambda.loc[i, 'lambda_'+str(x)] = topics_lambda_list
```
